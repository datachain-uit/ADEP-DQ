# Chia dữ liệu thành 4 phase dựa trên cột ['days_since_enroll']

## Hàm gán phase

```python
def assign_phase(days):
    if pd.isna(days):
        return None
    if days <= 14:
        return 'Phase_1'
    elif days <= 28:
        return 'Phase_2'
    elif days <= 42:
        return 'Phase_3'
    else:
        return 'Phase_4'
```

## Thực hiện gán phase

```python
# Đọc file
print("Đang đọc merged_df.csv...")
df = pd.read_csv("/kaggle/working/merged_df.csv")
print(f"Số dòng ban đầu: {len(df)}")

# Gán giá trị phase cho từng dòng dựa vào số ngày kể từ khi ghi danh
print("Đang gán phase...")
df['phase'] = df['days_since_enroll'].apply(assign_phase)  # Hàm assign_phase sẽ phân loại phase (1, 2, 3, 4)

df = df[df['phase'].notna()]
print(f"Số dòng sau khi gán phase: {len(df)}")

print("\nPhân bố phase:")
print(df['phase'].value_counts().sort_index())  # Đếm số dòng theo từng phase
```

## Kiểm tra trùng và groupby

```python
phases = ['Phase_1', 'Phase_2', 'Phase_3', 'Phase_4']
group_cols = ['user_id', 'course_id', 'enroll_time']

agg_results = {}

for i, phase in enumerate(phases, start=1):
    print(f"\n=== {phase} ===")

    df_phase = df[df['phase'] == phase].copy()

    # Bỏ cột không cần
    drop_cols = ['label_encoded', 'is_first_reply', 'days_since_enroll']
    df_phase = df_phase.drop(columns=drop_cols, errors='ignore')

    # Kiểm tra trùng
    num_dups_before = df_phase.duplicated(subset=group_cols, keep=False).sum()
    print(f"Trước khi gộp: {num_dups_before} bản ghi trùng dựa trên {group_cols}.")

    # Cột numeric (ngoại trừ 'phase')
    numeric_cols = df_phase.select_dtypes(include=['int64', 'float64']).columns.tolist()
    numeric_cols = [c for c in numeric_cols]

    # Tổng và trung bình
    sum_cols = ['num_words']
    mean_cols = [c for c in numeric_cols if c not in sum_cols]

    agg_dict = {col: 'mean' for col in mean_cols}
    for c in sum_cols:
        agg_dict[c] = 'sum'

    # Đếm sentiment
    sentiment_counts = (
        df_phase
        .groupby(group_cols + ['label'])
        .size()
        .unstack(fill_value=0)
        .rename(columns=lambda x: f"total_{x.lower()}")
        .reset_index()
    )

    # Gộp numeric
    agg_numeric = (
        df_phase
        .groupby(group_cols, as_index=False)
        .agg(agg_dict)
        .sort_values(by=group_cols)
        .reset_index(drop=True)
    )

    # Merge 2 phần
    agg_df = pd.merge(agg_numeric, sentiment_counts, on=group_cols, how='left')

    # Đảm bảo luôn có 3 cột này trong file
    for c in ['total_negative', 'total_neutral', 'total_positive']:
        if c not in agg_df.columns:
            agg_df[c] = 0

    # Gán số phase (1, 2, 3, 4)
    agg_df['phase'] = i
    agg_results[i] = agg_df

print("\nHoàn tất gộp dữ liệu tất cả phase.")
```

```python
# Kiểm tra dữ liệu sau groupby
for phase, agg_df in agg_results.items():
    num_dups_after = agg_df.duplicated(subset=group_cols, keep=False).sum()
    print(f"\n{phase}: Còn {num_dups_after} bản ghi trùng sau khi gộp.")
    
    if num_dups_after != 0:
        print("Vẫn còn trùng, cần kiểm tra logic groupby.")
```

## Lưu file

```python
# Lưu phase vào 4 file
for phase, agg_df in agg_results.items():
    filename = f"reply_phase_{phase}.csv"
    agg_df.to_csv(filename, index=False)
    print(f"Đã lưu {filename}: {len(agg_df)} dòng, {len(agg_df.columns)} cột")

print("\nHoàn tất xử lý và lưu tất cả 4 phase.")
```
