```python
import pandas as pd
```

```python
label = pd.read_csv(label_path)
label.head()
```

```python
user_problem = pd.read_parquet(user_problem_path)
user_problem.info()
```

```python
selected_label_col = ['user_id', 'course_id',	'enroll_time','enrollment_to_end']
selected_label = label[selected_label_col]
selected_label.head()
```

```python
merge_col = ['course_id', 'user_id']
merge = pd.merge(selected_label, user_problem, on=merge_col, how='inner')
merge.head()
```

```python
merge.shape
```

```python
merge.info()
```

```python
merge['submit_time'] = pd.to_datetime(merge['submit_time'])
merge['enroll_time'] = pd.to_datetime(merge['enroll_time'])
merge['duration_days'] = (merge['submit_time'] - merge['enroll_time']).dt.days
display(merge.head())
```

```python
import numpy as np

merge_filtered = merge.copy()

# Chuyển type từ one-hot
type_cols = ['type_1', 'type_2', 'type_3', 'type_4', 'type_5', 'type_6']
merge_filtered['type'] = merge_filtered[type_cols].dot(range(1, 7))

# Tính tỷ lệ thời gian = (ngày đã học / tổng thời gian)
merge_filtered['time_ratio'] = merge_filtered['duration_days'] / merge_filtered['enrollment_to_end']

# Điều kiện để gán phase theo tỷ lệ
conditions = [
    merge_filtered['time_ratio'] <= 0.25,
    merge_filtered['time_ratio'] <= 0.50,
    merge_filtered['time_ratio'] <= 0.75,
    merge_filtered['time_ratio'] <= 0.9
]
choices = [1, 2, 3, 4]

merge_filtered['phase'] = np.select(conditions, choices, default=None)

display(merge_filtered[['duration_days', 'enrollment_to_end', 'time_ratio', 'phase']].head())
```

```python
# Loại bỏ những dòng có phase none
merge_filtered = merge_filtered.dropna(subset=['phase'])
merge_filtered.shape
```

```python
merge_filtered['phase'].value_counts()
```

```python
merge_phase_1 = merge_filtered[merge_filtered['phase'] == 1]
merge_phase_2 = merge_filtered[merge_filtered['phase'] == 2]
merge_phase_3 = merge_filtered[merge_filtered['phase'] == 3]
merge_phase_4 = merge_filtered[merge_filtered['phase'] == 4]
```

```python
import numpy as np
group_by_col = ['user_id', 'course_id', 'exercise_id']

def calculate_average_time_between_submissions(submit_times):
    if len(submit_times) > 1:
        sorted_times = sorted(submit_times)
        diffs = [(sorted_times[i] - sorted_times[i-1]).total_seconds() for i in range(1, len(sorted_times))]
        return sum(diffs) / len(diffs)
    return 0

phases = [merge_phase_1, merge_phase_2, merge_phase_3, merge_phase_4]
grouped_results = []

for df in phases:
    df.loc[:, 'submit_time'] = pd.to_datetime(df['submit_time'])
    df.loc[:, 'enroll_time'] = pd.to_datetime(df['enroll_time'])
    grouped = df.groupby(group_by_col).agg(
        sum_attempts=('attempts', 'sum'),
        sum_score=('score', 'sum'),
        sum_max_score=('max_score_for_problem', 'sum'),
        submit_times=('submit_time', list),
        average_time_between_submissions=('submit_time', calculate_average_time_between_submissions),
        min_date_from_enrol_to_submit=('submit_time', 'min'),
        max_date_from_enrol_to_submit=('submit_time', 'max'),
        enroll_time=('enroll_time', 'first')
    ).reset_index()

    grouped['score_to_max_score_ratio'] = grouped['sum_score'] / grouped['sum_max_score'].replace(0, np.nan)
    grouped['enroll_to_first_submit_days'] = (grouped['min_date_from_enrol_to_submit'] - grouped['enroll_time']).dt.days
    grouped['enroll_to_last_submit_days'] = (grouped['max_date_from_enrol_to_submit'] - grouped['enroll_time']).dt.days
    grouped_results.append(grouped)

merge_phase_1_grouped, merge_phase_2_grouped, merge_phase_3_grouped, merge_phase_4_grouped = grouped_results
```

```python
def calculate_exercise_time(submit_times):
    if len(submit_times) > 0:
        return (max(submit_times) - min(submit_times)).total_seconds()
    else:
        return 0

merge_phase_1_grouped['exercise_time'] = merge_phase_1_grouped['submit_times'].apply(calculate_exercise_time)
merge_phase_2_grouped['exercise_time'] = merge_phase_2_grouped['submit_times'].apply(calculate_exercise_time)
merge_phase_3_grouped['exercise_time'] = merge_phase_3_grouped['submit_times'].apply(calculate_exercise_time)
merge_phase_4_grouped['exercise_time'] = merge_phase_4_grouped['submit_times'].apply(calculate_exercise_time)

print(merge_phase_1_grouped.shape)
print(merge_phase_2_grouped.shape)
print(merge_phase_3_grouped.shape)
print(merge_phase_4_grouped.shape)
```

```python
group_by_user_course = ['user_id', 'course_id']

merge_phase_1_aggregated = merge_phase_1_grouped.groupby(group_by_user_course).agg(
    average_attempts=('sum_attempts', 'mean'),
    sum_exercise=('exercise_id', 'count'),
    average_ratio=('score_to_max_score_ratio', 'mean'),
    average_exercise_time=('exercise_time', 'mean'),
    average_time_between_submissions=('average_time_between_submissions', 'mean'),
    average_enroll_to_first_submit_days=('enroll_to_first_submit_days', 'mean'),
    average_enroll_to_last_submit_days=('enroll_to_last_submit_days', 'mean')
).reset_index()

merge_phase_2_aggregated = merge_phase_2_grouped.groupby(group_by_user_course).agg(
    average_attempts=('sum_attempts', 'mean'),
    sum_exercise=('exercise_id', 'count'),
    average_ratio=('score_to_max_score_ratio', 'mean'),
    average_exercise_time=('exercise_time', 'mean'),
    average_time_between_submissions=('average_time_between_submissions', 'mean'),
    average_enroll_to_first_submit_days=('enroll_to_first_submit_days', 'mean'),
    average_enroll_to_last_submit_days=('enroll_to_last_submit_days', 'mean')
).reset_index()

merge_phase_3_aggregated = merge_phase_3_grouped.groupby(group_by_user_course).agg(
    average_attempts=('sum_attempts', 'mean'),
    sum_exercise=('exercise_id', 'count'),
    average_ratio=('score_to_max_score_ratio', 'mean'),
    average_exercise_time=('exercise_time', 'mean'),
    average_time_between_submissions=('average_time_between_submissions', 'mean'),
    average_enroll_to_first_submit_days=('enroll_to_first_submit_days', 'mean'),
    average_enroll_to_last_submit_days=('enroll_to_last_submit_days', 'mean')
).reset_index()

merge_phase_4_aggregated = merge_phase_4_grouped.groupby(group_by_user_course).agg(
    average_attempts=('sum_attempts', 'mean'),
    sum_exercise=('exercise_id', 'count'),
    average_ratio=('score_to_max_score_ratio', 'mean'),
    average_exercise_time=('exercise_time', 'mean'),
    average_time_between_submissions=('average_time_between_submissions', 'mean'),
    average_enroll_to_first_submit_days=('enroll_to_first_submit_days', 'mean'),
    average_enroll_to_last_submit_days=('enroll_to_last_submit_days', 'mean')
).reset_index()

display(merge_phase_1_aggregated.head())
display(merge_phase_2_aggregated.head())
display(merge_phase_3_aggregated.head())
display(merge_phase_4_aggregated.head())
```

```python
resource_exercise_path = '/content/drive/MyDrive/MOOCCubeX/Dataset/Data_cleaned/Exercise/resource_exercise_course_have_problem_info.parquet'
resource_exercise = pd.read_parquet(resource_exercise_path)
resource_exercise.head()
```

```python
merge_phase_1_merged = pd.merge(merge_phase_1_aggregated, resource_exercise[['id', 'ex_counts']], left_on='course_id', right_on='id', how='inner')
merge_phase_2_merged = pd.merge(merge_phase_2_aggregated, resource_exercise[['id', 'ex_counts']], left_on='course_id', right_on='id', how='inner')
merge_phase_3_merged = pd.merge(merge_phase_3_aggregated, resource_exercise[['id', 'ex_counts']], left_on='course_id', right_on='id', how='inner')
merge_phase_4_merged = pd.merge(merge_phase_4_aggregated, resource_exercise[['id', 'ex_counts']], left_on='course_id', right_on='id', how='inner')

merge_phase_1_merged['exercise_completion_ratio'] = merge_phase_1_merged['sum_exercise'] / merge_phase_1_merged['ex_counts']
merge_phase_2_merged['exercise_completion_ratio'] = merge_phase_2_merged['sum_exercise'] / merge_phase_2_merged['ex_counts']
merge_phase_3_merged['exercise_completion_ratio'] = merge_phase_3_merged['sum_exercise'] / merge_phase_3_merged['ex_counts']
merge_phase_4_merged['exercise_completion_ratio'] = merge_phase_4_merged['sum_exercise'] / merge_phase_4_merged['ex_counts']

display(merge_phase_1_merged.head())
display(merge_phase_2_merged.head())
display(merge_phase_3_merged.head())
display(merge_phase_4_merged.head())
```

```python
# Define the columns to rename
cols_to_rename = [
    'average_attempts',
    'average_ratio',
    'average_exercise_time',
    'average_time_between_submissions',
    'average_enroll_to_first_submit_days',
    'average_enroll_to_last_submit_days',
    'exercise_completion_ratio'
]

# Rename columns for each phase
rename_mapping_1 = {col: f'{col}_1' for col in cols_to_rename}
rename_mapping_2 = {col: f'{col}_2' for col in cols_to_rename}
rename_mapping_3 = {col: f'{col}_3' for col in cols_to_rename}
rename_mapping_4 = {col: f'{col}_4' for col in cols_to_rename}

merge_phase_1_merged.rename(columns=rename_mapping_1, inplace=True)
merge_phase_2_merged.rename(columns=rename_mapping_2, inplace=True)
merge_phase_3_merged.rename(columns=rename_mapping_3, inplace=True)
merge_phase_4_merged.rename(columns=rename_mapping_4, inplace=True)
```

```python


drop_col = ['sum_exercise', 'id', 'ex_counts']
merge_phase_1_merged.drop(drop_col, axis=1, inplace=True)
merge_phase_2_merged.drop(drop_col, axis=1, inplace=True)
merge_phase_3_merged.drop(drop_col, axis=1, inplace=True)
merge_phase_4_merged.drop(drop_col, axis=1, inplace=True)

display(merge_phase_1_merged.head())
display(merge_phase_2_merged.head())
display(merge_phase_3_merged.head())
display(merge_phase_4_merged.head())
```
