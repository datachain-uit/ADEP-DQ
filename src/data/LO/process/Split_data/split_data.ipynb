```python
import polars as pl

def process_date_columns(df: pl.DataFrame):
    if "start_date" not in df.columns:
        return df

    # 1) Convert start_date → Date
    df = df.with_columns(
        pl.col("start_date").str.strptime(pl.Date, strict=False)
    )

    # 2) Tạo các cột mới
    df = df.with_columns([
        pl.col("start_date").dt.year().alias("start_year"),
        pl.col("start_date").dt.month().alias("start_month"),
        pl.col("start_date").dt.day().alias("start_day"),
    ])

    # 3) Sắp xếp lại cột: đưa 3 cột mới ngay sau start_date
    cols = df.columns
    idx = cols.index("start_date")

    new_order = (
        cols[:idx+1] +                 # giữ các cột trước start_date
        ["start_year", "start_month", "start_day"] +  # chèn 3 cột mới
        [c for c in cols if c not in ["start_year", "start_month", "start_day"] and c not in cols[:idx+1]]
    )

    df = df.select(new_order)
    return df

df = process_date_columns(polars_df)
df = df.drop('start_date')
```

```python
import polars as pl


def label_encode_replace(df: pl.DataFrame, columns: list[str]):
    for col in columns:
        # 1) Lấy giá trị unique và sort để mapping ổn định
        unique_vals = df[col].unique().sort()

        # 2) Tạo mapping: string → số
        mapper = {val: i for i, val in enumerate(unique_vals)}

        # 3) Replace + ép kiểu số
        df = df.with_columns(
            pl.col(col)
            .replace(mapper)        # thay string thành số
            .cast(pl.Int64)         # ép về dạng numeric int
            .alias(col)
        )
    return df

df = label_encode_replace(df, ["name", "school_name", "user_id", "course_id", "label_3"])
```

```python
string_cols = [c for c, dtype in df.schema.items() if dtype == pl.Utf8]
print("String columns:", string_cols)

df = df.with_columns([
    pl.col(c).cast(pl.Float64).alias(c) for c in string_cols
])
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

pandas_df = df.to_pandas()

train_df, temp_df = train_test_split(
    pandas_df,
    test_size=0.2,
    stratify=pandas_df['label_3'],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df['label_3'],
    random_state=42
)

print(f"Train set shape: {train_df.shape}")
print(f"Validation set shape: {val_df.shape}")
print(f"Test set shape: {test_df.shape}")
```

```python
import pandas as pd

phase_end_columns = {
    'p1': 'cmt_p1_sentiment_entropy_phase1',
    'p2': 'cmt_p2_sentiment_entropy_phase2',
    'p3': 'cmt_p3_sentiment_entropy_phase3',
    'p4': 'cmt_p4_sentiment_entropy_phase4'
}

# Get the full list of columns from the validation DataFrame
all_columns = val_df.columns.tolist()

# Dictionaries to store the phase-specific DataFrames
val_dfs_by_phase = {}
test_dfs_by_phase = {}

print("Creating phase-specific DataFrames and zeroing out subsequent columns...")

for phase_name, end_col in phase_end_columns.items():
    if end_col in all_columns:
        # Find the index of the end column for the current phase
        end_col_idx = all_columns.index(end_col)

        # Get the names of columns that come *after* the end_col for this phase
        # These are the columns that should be zeroed out
        cols_to_zero_out = all_columns[end_col_idx + 1:]

        # Create a copy of the *entire* original DataFrame for each phase
        # and then set the 'future' columns to 0
        current_val_df_phase = val_df.copy()
        current_test_df_phase = test_df.copy()

        # Set the columns to zero
        for col in cols_to_zero_out:
            current_val_df_phase[col] = 0.0 # Use float to be consistent with existing numerical data
            current_test_df_phase[col] = 0.0

        val_dfs_by_phase[phase_name] = current_val_df_phase
        test_dfs_by_phase[phase_name] = current_test_df_phase

        print(f"  - {phase_name.upper()} Validation DataFrame shape: {val_dfs_by_phase[phase_name].shape}")
        print(f"  - {phase_name.upper()} Test DataFrame shape: {test_dfs_by_phase[phase_name].shape}")
    else:
        print(f"Warning: End column '{end_col}' for {phase_name.upper()} not found. Skipping phase.")

print("\nExample: Head of P1 Validation DataFrame (showing zeroed columns):")
display(val_dfs_by_phase['p1'].head())
```

```python
import os

# Define the output directory
output_dir = 'data_output'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

print(f"Saving DataFrames to CSV files in '{output_dir}'...")

# Save the training DataFrame
train_df_path = os.path.join(output_dir, 'train.csv')
train_df.to_csv(train_df_path, index=False)
print(f"  - Saved train_df to {train_df_path}")

# Save validation phase DataFrames
for phase_name, df_phase in val_dfs_by_phase.items():
    val_path = os.path.join(output_dir, f'val_{phase_name}.csv')
    df_phase.to_csv(val_path, index=False)
    print(f"  - Saved validation {phase_name.upper()} DataFrame to {val_path}")

# Save test phase DataFrames
for phase_name, df_phase in test_dfs_by_phase.items():
    test_path = os.path.join(output_dir, f'test_{phase_name}.csv')
    df_phase.to_csv(test_path, index=False)
    print(f"  - Saved test {phase_name.upper()} DataFrame to {test_path}")

print("All specified DataFrames have been saved.")
```
