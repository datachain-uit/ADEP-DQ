## Thực hiện gán phase

```python
df = merged_df.copy()
print(f"Số dòng ban đầu: {len(df)}")
# Hàm gán phase
def assign_phase(row):
    days = row['days_since_enroll']
    course_duration = row['enrollment_to_end']
    if days <= 0.25 * course_duration:
        return 'Phase_1'
    elif days <= 0.50 * course_duration:
        return 'Phase_2'
    elif days <= 0.75 * course_duration:
        return 'Phase_3'
    elif days <= 0.9 * course_duration:
        return 'Phase_4'
    else:
        return None

# Gán giá trị phase cho từng dòng dựa vào số ngày kể từ khi ghi danh và thời lượng khóa học
df['phase'] = df.apply(assign_phase, axis=1)

df = df[df['phase'].notna()]
print(f"Số dòng sau khi gán phase: {len(df)}")

print("\nPhân bố phase:")
print(df['phase'].value_counts().sort_index())  # Đếm số dòng theo từng phase
```

## Kiểm tra trùng lắp và groupby

```python
import pandas as pd

# ===== Hàm tính tổng thời gian xem thực tế (loại trừ chồng lấn) =====
def total_watch_time(start_point, end_point):
    intervals = sorted(zip(start_point, end_point), key=lambda x: x[0])
    merged = []
    for start, end in intervals:
        if not merged or start > merged[-1][1]:
            merged.append([start, end])
        else:
            merged[-1][1] = max(merged[-1][1], end)
    total = sum(e - s for s, e in merged)
    return total


# ===== Hàm tính các đặc trưng xem video =====
def compute_features(group):
    avg_segment_duration = (group['end_point'] - group['start_point']).mean()
    real_watch_time = total_watch_time(group['start_point'].tolist(), group['end_point'].tolist())
    avg_real_watch_time = real_watch_time / len(group)
    segment_count = len(group)
    avg_playback_speed = group['playback_speed'].mean()
    avg_day_of_week = round(group['day_of_week'].mean())

    most_common_hour = (
        group['hour_of_day'].value_counts().idxmax()
        if not group['hour_of_day'].isnull().all()
        else None
    )
    most_common_day = (
        group['day_of_week'].value_counts().idxmax()
        if not group['day_of_week'].isnull().all()
        else None
    )

    unique_video_count = group['video_id'].nunique()
    avg_segment_count_per_video = segment_count / unique_video_count if unique_video_count > 0 else 0
    var_playback_speed = group['playback_speed'].var(ddof=0)
    fast_forward_ratio = (group['playback_speed'] > 1.0).mean()

    if 'segment_duration_seconds' in group.columns:
        session_duration_std = group['segment_duration_seconds'].std(ddof=0)
    else:
        session_duration_std = (group['end_point'] - group['start_point']).std(ddof=0)

    return pd.Series({
        'avg_segment_duration': avg_segment_duration,
        'avg_real_watch_time': avg_real_watch_time,
        'segment_count': segment_count,
        'avg_playback_speed': avg_playback_speed,
        'avg_day_of_week': avg_day_of_week,
        'most_common_hour': most_common_hour,
        'most_common_day': most_common_day,
        'unique_video_count': unique_video_count,
        'avg_segment_count_per_video': avg_segment_count_per_video,
        'var_playback_speed': var_playback_speed,
        'fast_forward_ratio': fast_forward_ratio,
        'session_duration_std': session_duration_std
    })


# ===== Pipeline tổng hợp dữ liệu theo phase =====
phases = ['Phase_1', 'Phase_2', 'Phase_3', 'Phase_4']
group_cols = ['user_id', 'course_id', 'enroll_time']

agg_results = {}

for i, phase in enumerate(phases, start=1):
    print(f"\n=== {phase} ===")

    # Lọc dữ liệu theo phase
    df_phase = df[df['phase'] == phase].copy()

    # Bỏ cột không cần thiết
    drop_cols = ['days_since_enroll']
    df_phase = df_phase.drop(columns=drop_cols, errors='ignore')

    # Kiểm tra trùng lặp
    num_dups_before = df_phase.duplicated(subset=group_cols, keep=False).sum()
    print(f"Trước khi gộp: {num_dups_before} bản ghi trùng dựa trên {group_cols}.")

    if num_dups_before == 0:
        print("Không có bản ghi trùng, giữ nguyên dữ liệu.")
        df_phase['phase'] = i
        df_phase['num_videos_watched'] = 1
        agg_results[i] = df_phase.copy()

    else:
        print("Có trùng, tiến hành groupby và tính đặc trưng hành vi xem video...")

        # Tính đặc trưng chi tiết
        feature_df = df_phase.groupby(group_cols).apply(compute_features).reset_index()

        # Gộp thêm các đặc trưng numeric khác (ví dụ điểm, tỉ lệ, v.v.)
        numeric_cols = df_phase.select_dtypes(include=['int64', 'float64']).columns.tolist()
        numeric_cols = [c for c in numeric_cols if c not in ['phase', 'start_point', 'end_point']]

        sum_cols = ['segment_duration_seconds']
        mean_cols = [c for c in numeric_cols if c not in sum_cols]
        agg_dict = {col: 'mean' for col in mean_cols}
        for c in sum_cols:
            agg_dict[c] = 'sum'

        agg_numeric = (
            df_phase.groupby(group_cols, as_index=False)
            .agg(agg_dict)
            .sort_values(by=group_cols)
            .reset_index(drop=True)
        )

        # Đếm số video duy nhất
        video_count = (
            df_phase.groupby(group_cols)['video_id']
            .nunique()
            .reset_index(name='num_videos_watched')
        )

        # Gộp tất cả: numeric + feature + video_count
        agg_final = (
            agg_numeric
            .merge(feature_df, on=group_cols, how='left')
            .merge(video_count, on=group_cols, how='left')
        )

        agg_final['phase'] = i
        agg_results[i] = agg_final

print("\n✅ Hoàn tất gộp dữ liệu tất cả phase.")
```

```python
# Kiểm tra dữ liệu sau groupby
for phase, agg_df in agg_results.items():
    num_dups_after = agg_df.duplicated(subset=group_cols, keep=False).sum()
    print(f"\n{phase}: Còn {num_dups_after} bản ghi trùng sau khi gộp.")

    if num_dups_after != 0:
        print("Vẫn còn trùng, cần kiểm tra logic groupby.")
```

```python
cols_to_drop = [
    'enroll_time', 'Unnamed: 0', 'video', 'assignment', 'exam', 'amount_exercise_each_exam',
       'total_correct_ratio_exam', 'average_correct_ratio_exam',
       'amount_exercise_each_assignment', 'total_correct_ratio_assignment',
       'average_correct_ratio_assignment', 'watched_videos', 'video_counts',
       'watch_percent', 'score', 'course_duration', 'enrollment_to_end','playback_speed', 'timestamp',
    'hour_of_day', 'day_of_week','num_videos_watched'
]

for phase, agg_df in agg_results.items():
    agg_results[phase] = agg_df.drop(columns=cols_to_drop, errors='ignore')
    print(f"Columns in Phase {phase} after dropping: {agg_results[phase].columns.tolist()}")

print("\nFinished dropping columns from all phases.")
```

```python
for phase, agg_df in agg_results.items():
    # Rename columns by adding phase suffix, excluding 'user_id' and 'course_id'
    renamed_columns = {}
    for col in agg_df.columns:
        if col not in ['user_id', 'course_id']:
            renamed_columns[col] = f"{col}_{phase}"
        else:
            renamed_columns[col] = col  # Keep 'user_id' and 'course_id' as they are
    agg_results[phase] = agg_df.rename(columns=renamed_columns)

    print(f"Columns in Phase {phase} after renaming: {agg_results[phase].columns.tolist()}")

print("\nFinished renaming columns in all phases.")
```

## Lưu file

```python
# Lưu phase vào 4 file
for phase, agg_df in agg_results.items():
    # Drop the phase column before saving
    if f'phase_{phase}' in agg_df.columns:
      agg_df = agg_df.drop(columns=[f'phase_{phase}'])
    elif 'phase' in agg_df.columns: # Handle case where phase column was not renamed
      agg_df = agg_df.drop(columns=['phase'])

    filename = f"/content/drive/MyDrive/MOOCCubeX/Dataset/Data_cleaned/video_phase_{phase}.csv"
    agg_df.to_csv(filename, index=False)
    print(f"Đã lưu {filename}: {len(agg_df)} dòng, {len(agg_df.columns)} cột")

print("\nHoàn tất xử lý và lưu tất cả 4 phase.")
```
